```python
#importamos la libreria.
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# leemos la imagen a filtrar y la cambiamos a color blanco y negro
img = cv.imread('guason.jpg', 0)

# obtenemos las dimensiones de la imagen
row, column = img.shape

# devuelve un array del tamaño de filas y columnas.
img_filter = np.zeros([row, column])

# iniciamos el ciclo for para compensar el ruido que tiene la imagen
for i in range(1, row - 1):
    for j in range(1, column - 1):
        temp = [img[i - 1, j + 1],
                img[i + 1, j - 1],
                img[i - 1, j],
                img[i, j - 1],
                img[i, j],
                img[i, j + 1],
                img[i + 1, j + 1],
                img[i, j],
                img[i + 1, j - 1],
                img[i + 1, j],
                img[i, j - 1],
                img[i + 1, j + 1]]
        temp = sorted(temp)
        img_filter[i, j] = temp[5]

# cambiamos la imagen a escala de blanco y negro para la visualización en el sistema.
img_filter = img_filter.astype(np.uint8)
img_gray = cv.cvtColor(img, cv.COLOR_GRAY2BGRA)
img_gray1 = cv.cvtColor(img_filter, cv.COLOR_GRAY2BGRA)

# acomodamos las imágenes para poder visualizarlas de mejor manera.
plt.figure(figsize=(10, 10), dpi=100)
plt.subplot(121), plt.imshow(img_gray[:, ::1]), plt.title('Imagen Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_gray1[:, ::1]), plt.title('Imagen Filtrada')
plt.xticks([]), plt.yticks([])
plt.show()
```
